```python
# Imports
from ultralytics import YOLO
```

```python
# Model
model = YOLO("yolov8n.pt")
```

```python
# Treinando com dados de só 1 minuto
results = model.train(
    data="data.yaml", 
    epochs=10,
    imgsz=640,
    batch=16,
    name="casemiro-train",
)
```

```python
# Copiando o modelo treinado acima
import shutil
import os

os.makedirs("models", exist_ok=True)

shutil.copy(
    "runs/detect/casemiro-train/weights/best.pt",
    "models/min1.pt"
)

# Usando o modelo treinado acima para testar
model_min1 = YOLO("models/min1.pt")
model_min1.track(
    source="Primeiro tempo leve.mp4",
    tracker="botsort.yaml",
    conf=0.3,
    iou=0.5,
    persist=True,
    save=True,
    show=True
)
```

```python
# Treinando com dados do min 1 ao 2
model2 = YOLO("runs/detect/casemiro-train/weights/last.pt")

results = model2.train(
    data="data.yaml",
    epochs=10,
    imgsz=640,
    batch=16,
    name="casemiro-train",
    exist_ok=True
)
```
